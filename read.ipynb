```python
import pandas as pd
df = pd.read_csv("student_dropout.csv")
inschool = df[df["Dropped_Out"] == False ]
inschool
droppedout = df[df["Dropped_Out"] == True]
droppedout
#droppedout["total"] = len(droppedout)
age = droppedout.groupby("Age").agg("count").reset_index()
#1 - Age
ageplot1= age.plot.bar(x="Age",y="Dropped_Out")
#2- Gender
genderplot= droppedout.groupby("Gender").agg("count").reset_index()
genderplot1 = genderplot.plot.bar(x="Gender",y="Dropped_Out")
 #3 - Parental Status 
parentalplot= droppedout.groupby("Parental_Status").agg("count").reset_index()
parentalplot1 = parentalplot.plot.bar(x="Parental_Status",y="Dropped_Out")
#4- Weekday Alcohol Constumption
wacplot = droppedout.groupby("Weekday_Alcohol_Consumption").agg("count").reset_index()
wacplot
wacplot1 = wacplot.plot.bar(x="Weekday_Alcohol_Consumption",y="Dropped_Out")

#5 - Family Support 
#Dropped Out
famSupport = droppedout.groupby("Family_Support").agg("count").reset_index()
famSupport1 = famSupport.plot.bar(x = "Family_Support", y = "Dropped_Out", title = "Family Support for Students Who Dropped Out" )
#In School
famSupport = inschool.groupby("Family_Support").agg("count").reset_index()
famSupport2 = famSupport.plot.bar(x = "Family_Support", y = "Dropped_Out" , title = "Family Support for Students Who Stayed in School")

#6 = Internet Access
#Dropped Out
internetAcc = droppedout.groupby("Internet_Access").agg("count").reset_index()
internetAcc1 = internetAcc.plot.bar(x = "Internet_Access", y = "Dropped_Out", title = "Internet Access for Students Who Dropped Out")
#In School
internetAcc = inschool.groupby("Internet_Access").agg("count").reset_index()
internetAcc2 = internetAcc.plot.bar(x = "Internet_Access", y = "Dropped_Out", title = "Internet Access for Students Who Stayed in School")

#7 = Health Status
#Dropped Out
healthStat = droppedout.groupby("Health_Status").agg("count").reset_index()
healthStat1 = healthStat.plot.bar(x = "Health_Status", y = "Dropped_Out", title = "Health Rating of the Students Who Dropped Out")
#In School
healthStat = inschool.groupby("Health_Status").agg("count").reset_index()
healthStat2 = healthStat.plot.bar(x = "Health_Status", y = "Dropped_Out", title = "Health Rating of the Students Who Stayed in School")

#8 = Weekend Alcohol Consumption
#Dropped Out
weekendAl = droppedout.groupby("Weekend_Alcohol_Consumption").agg("count").reset_index()
weekendAl1 = weekendAl.plot.bar(x = "Weekend_Alcohol_Consumption", y = "Dropped_Out", title = "Weekend Alcohol Consumption of the Students Who Dropped Out")
#In School
weekendAl = inschool.groupby("Weekend_Alcohol_Consumption").agg("count").reset_index()
weekendAl2 = weekendAl.plot.bar(x = "Weekend_Alcohol_Consumption", y = "Dropped_Out", title = "Weekend Alcohol Consumption of the Students Who Stayed in School")
```

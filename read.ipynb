```python
import pandas as pd
df = pd.read_csv("student dropout.csv")
inschool = df[df["Dropped_Out"] == False ]
inschool
droppedout = df[df["Dropped_Out"] == True]
droppedout
droppedout["total"] = len(droppedout)
ageplot= droppedout.groupby("Age").agg("count").reset_index()
ageplot.plot.bar(x="Age",y="Dropped_Out")
```
